```python
# Bibliotheken abrufen
import json
import pandas as pd
import requests
import numpy as np
```

```python
# URL als Variable definieren
url = "https://rodare.hzdr.de/api/records/?page=1&sort=mostrecent&size=1000"
```

```python
# Funktion um Inhalt aus URL in r zu speichern
r = requests.get(url)
```

```python
# JSON lesen lassen
json = r.json()
```

```python
# ARRAY aufbauen um ???
data = np.zeros([json['hits']['total'],1])
```

```python
# Definition einer Gesamtliste
list_all = []
```

```python
# Definition einer Teilliste und for-Schleife, um den type der Records aus allen Einträgen auszulesen und aufzulisten
type_list = []
for index,record in enumerate(json['hits']['hits']):
    type_list.append(record['metadata']['resource_type']['type'])
```

```python
# Hinzufügen der Teilliste zu Gesamtliste
list_all.append(type_list)
```

```python
# Leerung der Teilliste und Füllung mit dem Publikationsdatum
type_list = []
for index,record in enumerate(json['hits']['hits']):
    type_list.append(record['metadata']['publication_date'])
```

```python
# Hinzufügen der Teilliste zu Gesamtliste
list_all.append(type_list)
```

```python
# Leerung der Teilliste und Füllung mit dem Zugriffsrecht für Records
type_list = []
for index,record in enumerate(json['hits']['hits']):
    type_list.append(record['metadata']['access_right'])
```

```python
# Hinzufügen der Teilliste zu Gesamtliste
list_all.append(type_list)
```

```python
# Leerung der Teilliste und Füllung mit den Gesamtdownloadzahlen für Records
type_list = []
for index,record in enumerate(json['hits']['hits']):
    type_list.append(record['stats']['downloads'])
```

```python
# Hinzufügen der Teilliste zu Gesamtliste
list_all.append(type_list)
```
